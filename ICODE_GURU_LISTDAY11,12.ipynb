Slicing

```python
fruits=["apple","mango","banana","pineapple"]
print(fruits[0:3])
```

```python
fruits=["apple","mango","banana","pineapple"]
print(fruits[::2])#backword
```

```python
fruits=["apple","mango","banana","pineapple"]
print(fruits[2::])##foward
```

```python
fruits=["apple","mango","banana","pineapple"]
for fruit in fruits[:3]:
   print(fruit.title())
```

```python
fruits=["apple","mango","banana","pineapple"]
for fruit in fruits[3::]:
   print(fruit.title())
```
